```python
from time import gmtime, strftime
import json
import math

price = 9.25
shares = 1081
dividend = 0.15
currYear = int(strftime("%Y"))
maxYears = 10

# price = 9.25
# shares = 1081
# dividend = 0.15
# currYear = int(strftime("%Y"))
# maxYears = 10


years = {}

# Create a history plan spanning 10 years of dividends
# In each year are 4 Quarters, Annual Income, and Annual Shares
# In each quarter there are D/P, Re, and Cur
# Note: Q/P = Quarterly Profit/Price, Re = ReInvest Stocks, Cur = Current Stocks
#       Q/p is calculated to find how many additional stocks to buy from profit
#       Re is rounded down from Q/P to estimate the amount of additional stocks
#       Cur is the current amount of stocks in that quarter after taking in Re
for yr in range(0,maxYears+1):
    years[currYear+yr] = {}
    AI = 0 # Annual Income
    
    for q in range(1,5):
        Q = {}
        Q["Profit"] = round(shares * dividend,3) # rounded down to cents
        AI += Q["Profit"]
        Q["QP"] = round(Q["Profit"]/price,3) # rounded down to 3 decs
        Q["Re"] = math.floor(Q["QP"])
        Q["Cur"] = shares
#         AS = {}
#         AS.Q["Cur"])
        shares += Q["Re"] # update the current shares for next quarter
        years[currYear+yr]["Q"+str(q)] = Q
    
    years[currYear+yr]["Annual Income"] = round(AI, 3)
    years[currYear+yr]["Annual Shares"] = years[currYear+yr]["Q4"]["Cur"]       
        

print(json.dumps(years, indent = 4))
```
